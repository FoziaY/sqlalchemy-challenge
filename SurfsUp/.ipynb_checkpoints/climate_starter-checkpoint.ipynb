# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine,inspect 
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine,reflect=True)
```

```python
Inspector = inspect(engine)
```

```python
# View all of the classes that automap found Inspector = inspect(engine)
Tablenames= Inspector.get_table_names()
Tablenames
```

```python
# Save references to each table
Class_Station = Base.classes.station
Class_Measurement = Base.classes.measurement
```

```python
Class_Station
```

```python
Class_Measurement
```

```python
# Create our session (link) from Python to the DB
session_link = Session(engine)
session_link 
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
Latest_date=session_link.query(Class_Measurement.date).order_by(Class_Measurement.date.desc()).first()[0]

print("the most recent date is",Latest_date)
```

```python
#checking the coloums respresnting the measurement  table and the stattion table  
Measurement_col = Inspector.get_columns("measurement")
print("Measurement Table")
for col in Measurement_col:
    print(col)
    
```

```python
Station_col = Inspector.get_columns("Station")
print("Station Table")
for col in Station_col:
    print(col)
```

```python
 #Design a query to retrieve the last 12 months of precipitation data and plot the results.
 # Starting from the most recent data point in the database.    
# Sort the dataframe by date# Sort the dataframe by date
Precipitation_data=session_link.query(Class_Measurement.date,Class_Measurement.prcp).\
    filter(Class_Measurement.date >'2016-08-23').\
    order_by(Class_Measurement.date.desc()).all()
```

```python
#calling Precipitation_data
# Sort the dataframe by date# Sort the dataframe by date_data
```

```python
# Save the query results as a Pandas DataFrame. Explicitly set the column names
Precipitation_df = pd.DataFrame(Precipitation_data,columns=["date","precipitation"])
Precipitation_df 
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
plt.figure(figsize=(9,6))
plt.bar(Precipitation_df["date"],Precipitation_df["precipitation"],label="Precipitation") 
plt.xlabel("Date")
plt.ylabel("Precipitation")
plt.title("12 month of Precipitation data")
plt.xticks(Precipitation_df['date'].iloc[::len(Precipitation_df)//4], rotation=45)
plt.legend()
plt.show()

```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
Precipitation_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
Total_station = session_link.query(func.count(Class_Station.station)).scalar()

print("Number of station is the data set is :",Total_station)
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
Obeservation_station = session_link.query(Class_Station.station,func.count(Class_Station.station)).\
group_by(Class_Station.station).order_by(func.count(Class_Station.station).desc()).all()
```

```python
Obeservation_station
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
for station,count in Obeservation_station:
    print("station is",station,"and obesevation count is",count)
```

```python
most_active_station = Obeservation_station[0][0]
most_active_station
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

Temperature_details = session_link.query(func.min(Class_Measurement.tobs),func.max(Class_Measurement.tobs),func.avg(Class_Measurement.tobs)).\
filter(Class_Measurement.station == most_active_station).all()
Temperature_details


```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

Temperature_data=session_link.query(Class_Measurement.date,Class_Measurement.tobs).\
    filter(Class_Measurement.date >'2016-08-23').\
    order_by(Class_Measurement.date.desc()).all()
```

```python
Temperature_data
```

```python
Temperature_df = pd.DataFrame(Temperature_data,columns=["Date","Temperature"])
```

```python
plt.figure(figsize=(9,6))
plt.hist(Temperature_df["Temperature"],bins=12,label="Tempersture observation",edgecolor ="black" )
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("12 month of Temperature observation")
plt.tight_layout()
plt.legend()
plt.show()
```
